# PANDAS :-

Pandas is a Python library used for data analysis and data manipulation. It’s
one of the most important tools for data analysts, data scientists, and machine learning engineers.


In short padas is use for the data visulatization.

Pandas helps you read, clean, analyze, and process data easily — just like Excel but with Python code.

Pandas is like Excel + SQL + Python — all combined in one tool for data work.

```python
# 1-D Series using list :-

import pandas as pd
x=pd.Series([1,2,3,4,5])
print(x)
```

```python
# 1-D Series using tuple :-

import pandas as pd
a=(1,2,3,4,5)
x=pd.Series(a)
print(x)
```

```python
import pandas as pd
x=pd.Series([1,2,3,4],index=['a','b','c','d'])
print(x)
```

```python
import pandas as pd
x=pd.Series([1,2,3,4],index=['a','b','c','d'],name='jai')
print(x)
```

```python
print(x['b'])
print(x[-1])
```

```python
import pandas as pd
x=pd.Series(range(10),name='Kajal')
print(x)
```

```python
import pandas as pd
x=pd.Series(range(1,11),name='Kajal')
print(x)
```

```python
import pandas as pd
x=pd.Series(range(1,10),name='Kajal',index=[x for x in 'abcdefghi'])
print(x)


# OR

import pandas as pd
x = pd.Series(range(1, 10), name='Kajal', index=list('abcdefghi'))
print(x)
```

# Series using NumPy Array :-

```python
import pandas as pd
import numpy as np
df=np.array([1,2,3,4,5])
print(df)

df1=pd.Series(df,name='Ajay')
print(df1)

```
