**Logical Operator**

# AND &

```python
df[(df['x']>0) & (df['w']>1)]
```

# Or |

```python
df[(df['x']>0) | (df['w']>1)]
```

# NOT !

Not with and { & }

```python
df[~((df['x'] > 0) & (df['w'] > 1))]
```

Not with | { or }

```python
df[~((df['x'] > 0) | (df['w'] > 1))]
```

```python
import pandas as pd
df1=pd.DataFrame({'Name':['a','b','c'],'Salary':[12000,13000,24000]})
print(df1)
df2=pd.DataFrame({'Name':['a','b','c'],'Age':[12,13,24]})
print(df2)
pd.merge(df1,df2,on='Name')
```

```python
import pandas as pd
df1=pd.DataFrame({'Name':['a','b','c'],'Salary':[12000,13000,24000]})
print(df1)
df2=pd.DataFrame({'Name':['a','d','e'],'Age':[12,13,24]})
print(df2)
pd.merge(df1,df2,on='Name')
```

```python
pd.merge(df1,df2,on='Name',how='inner')
```

```python
pd.merge(df1,df2,on='Name',how='left')
```

```python
pd.merge(df1,df2,on='Name',how='right')
```

```python
pd.merge(df1,df2,on='Name',how='outer')
```
