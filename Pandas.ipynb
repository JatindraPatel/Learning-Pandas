**Series with linspace Function**

# np.arange()
---- Create sequence with a step size

**Syntax**

----- np.arange(start, stop, step)

# np.linspace()
------ Create sequence with a fixed number of elements

**Syntax**

------ np.linspace(start, stop, num)

```python
import numpy as np
import pandas as pd
# series with numpy linspace()
ser1=pd.Series(np.linspace(3,33,3))
print(ser1)

# series with numpy linspace()
ser1=pd.Series(np.linspace(3,33,3),index=list('abc'))
print(ser1)

# series with numpy linspace()
ser3=pd.Series(np.linspace(3,33,4))
print(ser3)

# series with numpy linspace()
ser3=pd.Series(np.linspace(3,33,4),index=list('abcd'))
print(ser3)

# series with numpy linspace()
ser4=pd.Series(np.linspace(1,100,10))
print(ser4)

```

# Series from Dictionary :-

```python
print(df[df['x']>0].loc['b']['w'])
```

**Logical Operator**

# AND &

```python
df[(df['x']>0) & (df['w']>1)]
```

# Or |

```python
df[(df['x']>0) | (df['w']>1)]
```

# NOT !

Not with and { & }

```python
df[~((df['x'] > 0) & (df['w'] > 1))]
```

Not with | { or }

```python
df[~((df['x'] > 0) | (df['w'] > 1))]
```

```python
import numpy as np
import pandas as pd
from numpy.random import randn
outer=['a','a','a','b','b','b']
inner=[1,2,3,1,2,3]
h_index=list(zip(outer,inner))
h_index=pd.MultiIndex.from_tuples(h_index)
df1=pd.DataFrame(randn(6,2))
print(df1)
```

```python
df1=pd.DataFrame(randn(6,2),h_index)
print(df1)
```

```python
df1=pd.DataFrame(randn(6,2),h_index,['A','B'])
print(df1)
```
