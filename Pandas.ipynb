**Series with linspace Function**

# np.arange()
---- Create sequence with a step size

**Syntax**

----- np.arange(start, stop, step)

# np.linspace()
------ Create sequence with a fixed number of elements

**Syntax**

------ np.linspace(start, stop, num)

```python
import numpy as np
import pandas as pd
# series with numpy linspace()
ser1=pd.Series(np.linspace(3,33,3))
print(ser1)

# series with numpy linspace()
ser1=pd.Series(np.linspace(3,33,3),index=list('abc'))
print(ser1)

# series with numpy linspace()
ser3=pd.Series(np.linspace(3,33,4))
print(ser3)

# series with numpy linspace()
ser3=pd.Series(np.linspace(3,33,4),index=list('abcd'))
print(ser3)

# series with numpy linspace()
ser4=pd.Series(np.linspace(1,100,10))
print(ser4)

```

# Concat function

```python
import pandas as pd
ser1=pd.Series([2,4,6,8])
ser2=pd.Series([10,12,14,16])
ser3=pd.concat([ser1,ser2],axis=0)
print(ser3)
```

```python
import pandas as pd
ser1=pd.Series([2,4,6,8])
ser2=pd.Series([10,12,14,16])
ser3=pd.concat([ser1,ser2],axis=1)
print(ser3)
```
